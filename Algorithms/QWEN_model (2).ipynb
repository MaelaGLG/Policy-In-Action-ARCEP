```python
import torch
from transformers import AutoProcessor, Qwen2_5_VLForConditionalGeneration
from PIL import Image
import requests

# Charger le modèle sur le CPU ou le GPU s'il est disponible
device = "cuda" if torch.cuda.is_available() else "cpu"

model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2.5-VL-7B-Instruct", torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32, device_map="auto"
).to(device)

# Charger le processeur
processor = AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-7B-Instruct")

# Charger une image depuis une URL ou un fichier local
image_path = "data/image.png"  # Remplacez par le chemin réel
image = Image.open(image_path)
# image = Image.open("chemin/vers/image.jpg")  # Si depuis un fichier local

# Définir la question
question = "Que représente cette image ?"

# Préparer les entrées pour le modèle
inputs = processor(images=image, text=question, return_tensors="pt").to(device)

# Générer une réponse
with torch.no_grad():
    generated_ids = model.generate(**inputs, max_new_tokens=128)

# Décoder la réponse
response = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("Réponse :", response)
```
